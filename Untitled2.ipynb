```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
Univ = pd.read_csv("Universities.csv")
```

```python
Univ
```

```python
array=Univ.iloc[:,1:].values
```

```python
array
```

```python
scaler= StandardScaler()
univ_s=scaler.fit_transform(array)
```

```python
model=KMeans(4,random_state=45)
```

```python
cluster=model.fit(univ_s)
```

```python
cluster.labels_
```

```python
Univ['Cluster']=cluster.labels_
```

```python
Univ
```

```python
Univ.groupby(['Cluster']).agg(['mean']).reset_index()
```
